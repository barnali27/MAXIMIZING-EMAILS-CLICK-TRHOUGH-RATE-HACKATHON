### Univariate, Bivariate and Multivariate Analysis:

```python
def univariateAnalysis_numeric(column,nbins):
    print("Description of " + column)
    print("----------------------------------------------------------------------------")
    print(train[column].describe(),end="\n")
    print(" ")
    
    
    fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, sharex=False, figsize=(15,3));
    sns.histplot(data=train, x= train[column], kde=True, ax=ax1)
    ax1.tick_params(axis='x', labelrotation=90)
    sns.boxplot(x=train[column], data=train, ax=ax2)
    ax2.tick_params(axis='x', labelrotation=90);
```

```python
num = train.select_dtypes(include = ['float64', 'int64'])
lstnumericcolumns = list(num.columns.values)
len(lstnumericcolumns)
```

```python
for x in lstnumericcolumns:
    univariateAnalysis_numeric(x,20)
```

```python
def bivariateAnalysis(column):
    fig, (ax) = plt.subplots(ncols=1, nrows=1, sharex=False, figsize=(10,3));
    sns.barplot(data=train, x = train[column], y= 'click_rate', ci=None);    
    ax.tick_params(axis='x', labelrotation=90);
    ax.set_title("click rate vs " + column)
```

```python
num = train.select_dtypes(include = ['float64', 'int64'])
lstnumericcolumns = list(num.columns.values)
lst = lstnumericcolumns
```

```python
for x in lst:
    bivariateAnalysis(x)
```

```python
corr = num.corr(method='pearson')
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool)) 
fig = plt.subplots(figsize=(25, 15))
sns.heatmap(num.corr(), annot=True,fmt='.2f',mask=mask)  
plt.show()
```

### There is no significant Multicollinearity observed in the above heatmap.
